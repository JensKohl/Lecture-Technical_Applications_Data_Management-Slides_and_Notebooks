### Schritt 3.1: Deskriptive Analyse

Machen wir weiter mit der deskriptiven Analyse. 
Ein guter Einstieg ist der Befehl describe von Pandas. Wir lernen spter noch eine bessere Methode kennen, die vieles fr uns abdeckt (Pandas Profiling).

```python
# Verwenden Sie die Methode describe mit Parameter include='all auf Ihr Dataframe
IMDB_df.describe(include='all')
```

Wir knnen hier schon einige deskriptive Sachen ableiten:
* **Zeile Count**: wir haben in allen Spalten bis auf Content_rating 979 Werte, d.h. bei content_rating fehlen 3 Werte oder die Filme sind nicht bewertet
* **Zeile unique**: Es gibt 12 verschiedene ratings und 16 verschiedene Genres. 975 von 979 Titeln eindeutig, d.h. 4 Filme mit gleichem Namen.
* **Zeile Top**: was die hufigsten Werte sind in einer Spalte mit mehreren gleichen Eintrgen, z.B. Dracula bei title oder R bei rating.
* **Zeile Frequency**: wie hufig der top Eintrag vorkommt, bspw. 460 mal R rating
* **Zeile mean**: Mittelwert, z.B. 7,8 beim Star Rating
* **Zeile Std**: Standardabweichung, d.h. durchschnittliche Breite der Streuung einzelner Werte um den Mittelwert herum.
* **Zeile min**: geringster Wert, bspw. 7,4 fr Star rating
* **Zeile max**: hchster Wert, bspw. 9,3 fr Star rating
* **Zeile 50%**: dieser auch Median genannte Wert, der genau in der Mitte der Daten liegt, d.h. er teilt die Daten der Spalte in 2 genau gleich groe Untermengen (Beispiel: fr die Menge 1, 3, 4, 10, 15, 20, 40 ist 10 der Median, da links und rechts jeweils 3 Elemente sind). Bei einer Menge mit einer geraden Anzahl werden die "beiden Mitten" durch 2 geteilt und dieser Wert als Median genommen.
* **Zeile 25%**, das untere Quantil. Teilt die untere Teilmenge kleiner 50% wiederum in 2 gleich groe Mengen. Wir erhalten die Zahl, die die untere Teilmenge geteilt hat. Diese Zahl ist somit grer als 25% der Daten (Beispiel: aus der oberen Menge erhalten wir 1, 3, 4. Somit ist 3 der Median der unteren Menge und dadurch das 25% Quantil der gesamten Menge).
* **Zeile 75%**, das obere Quantil. Teilt die obere Teilmenge wiederum in 2 gleich groe Mengen. Wir erhalten einen Wert, der grer ist als 75% der Werte.

Die deskriptive Analyse hilft uns, schnell wesentliche Eigenschaften eines Datensatzes zu erkennen.

Probieren Sie auch aus, wie es ohne diesen Parameter aussieht.

```python
ax = sns.histplot(x="genre", data= IMDB_df, stat='percent');

ax.axes.set_title("Genre in Percentage", fontsize=20);
ax.set_xticklabels(ax.get_xticklabels(), rotation=45);
```

```python
ax = sns.histplot(x="content_rating", data= IMDB_df, stat='percent');

ax.axes.set_title("Rating in Percentage", fontsize=20);
ax.set_xticklabels(ax.get_xticklabels(), rotation=45);
```

```python
# Histogramm ber Star rating und Visualisierung mit Prozent.
ax = sns.histplot(x="star_rating", data= IMDB_df, stat="percent");
# fge einen Titel hinzu mit Schriftgre 20.
ax.axes.set_title("Star Rating in Percentage", fontsize=20);
# Style: wir drehen die Y-Labels um 30.
ax.set_yticklabels(ax.get_yticks(), size = 12, rotation=30);
```
